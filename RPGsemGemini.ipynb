```python
!pip install gemini-api

import google.generativeai as genai
import random

from google.colab import userdata

GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

# Função para gerar imagem com a API do Gemini
def gerar_imagem(prompt):
  response = genai.generate_image(prompt=prompt)
  return response['url']

# Função para batalha
def batalha(jogador, inimigo):
  print(f"You encountered a {inimigo['nome']}!")
  print(gerar_imagem(f"A {inimigo['nome']} ready for combat"))
  while jogador['vida'] > 0 and inimigo['vida'] > 0:
    print(f"Your Life: {jogador['vida']}, Enemy Life: {inimigo['vida']}")
    escolha = input("Choose your action (Attack, Defend, Use Item, Flee): ").lower()

    if escolha == "attack":
      dano = jogador['ataque'] - inimigo['defesa']
      if dano > 0:
        inimigo['vida'] -= dano
        print(f"You dealt {dano} damage to the {inimigo['nome']}!")
      else:
        print(f"Your attack did no damage to the {inimigo['nome']}.")
    elif escolha == "defend":
      print("You defended yourself from the enemy's attack.")
    elif escolha == "use item":
      print("You have no items.")
    elif escolha == "flee":
      print("You fled the battle!")
      return
    else:
      print("Invalid choice.")

    if inimigo['vida'] > 0:
      dano = inimigo['ataque'] - jogador['defesa']
      if dano > 0:
        jogador['vida'] -= dano
        print(f"The {inimigo['nome']} dealt {dano} damage to you!")
      else:
        print(f"The {inimigo['nome']} missed you.")

  if jogador['vida'] <= 0:
    print("You died!")
    print(gerar_imagem("Game Over"))
    return False
  else:
    print(f"You defeated the {inimigo['nome']}!")
    return True


# Aventuras predefinidas
aventuras = [
  {
    "titulo": "The Lost Sword of Eldoria",
    "introducao": "The kingdom of Eldoria is in peril. The legendary Sword of Eldoria, said to hold the power to banish all evil, has been stolen by the evil sorcerer Malkor. You must embark on a quest to retrieve the sword and save the kingdom.",
    "capitulos": [
      {
        "texto": "You arrive at the entrance of the Dark Forest, where Malkor is rumored to be hiding. The forest is shrouded in an eerie silence. You can hear the rustling of leaves and the occasional screech of an owl. What do you do?",
        "opcoes": ["* Enter the forest cautiously.", "* Search for another way into Malkor's lair.", "* Set up camp and wait for sunrise."]
      },
      {
        "texto": "As you venture deeper into the forest, you come across a fork in the path. Which way do you go?",
        "opcoes": ["* Take the left path, which seems to lead deeper into the forest.", "* Take the right path, which leads towards a faint light in the distance."]
      },
      {
        "texto": "You finally reach Malkor's lair, a dark and imposing fortress. What do you do?",
        "opcoes": ["* Charge into the fortress and confront Malkor.", "* Sneak into the fortress and search for the sword.", "* Try to find a way to weaken Malkor before confronting him."]
      }
    ],
    "conclusao": "You have faced many perils on your quest, but your courage and determination have led you to victory. The Sword of Eldoria is yours, and the kingdom is safe. Your name will be etched in the annals of Eldoria as a true hero."
  },
  # Adicione mais aventuras aqui...
]

# Função para iniciar o jogo
def iniciar_jogo():
  print("Welcome to the Adventure!")

  # Escolha da classe
  print("Choose your class:")
  print("1 - Warrior")
  print("2 - Mage")
  print("3 - Rogue")
  print("4 - Cleric")

  while True:
    escolha_classe = input("Your choice: ")
    if escolha_classe in ["1", "2", "3", "4"]:
      break
    else:
      print("Invalid choice.")

  classes = {
      "1": {"nome": "Warrior", "vida": 100, "ataque": 10, "defesa": 8},
      "2": {"nome": "Mage", "vida": 80, "ataque": 12, "defesa": 6},
      "3": {"nome": "Rogue", "vida": 90, "ataque": 8, "defesa": 10},
      "4": {"nome": "Cleric", "vida": 120, "ataque": 6, "defesa": 12}
  }
  jogador = classes[escolha_classe]
  print(f"You are now a {jogador['nome']}!")

  # Escolher Aventura Aleatória
  aventura_escolhida = random.choice(aventuras)

  print(aventura_escolhida['titulo'])
  print(aventura_escolhida['introducao'])

  # Iniciar a aventura capítulo por capítulo
  for i, capitulo in enumerate(aventura_escolhida['capitulos']):
    print(f"\n**Chapter {i+1}:**")
    print(capitulo['texto'])

    for j, opcao in enumerate(capitulo['opcoes']):
      print(f"{j+1} - {opcao[2:]}")

    while True:
      escolha = input("Your choice: ")
      if escolha.isdigit() and 1 <= int(escolha) <= len(capitulo['opcoes']):
        escolha = int(escolha)
        break
      else:
        print("Invalid choice.")

    # Implementar a lógica da escolha
    acao = capitulo['opcoes'][escolha - 1][2:].lower()
    if "attack" in acao:
      # Simular um combate
      inimigo = {
        "nome": "Goblin",
        "vida": 50,
        "ataque": 5,
        "defesa": 3
      }
      batalha(jogador, inimigo)
    elif "talk" in acao or "investigate" in acao or "use" in acao or "go" in acao:
      print("You chose to", acao)  # Implemente a lógica específica para cada ação

  print(f"\n{aventura_escolhida['conclusao']}")

# Iniciar o jogo
iniciar_jogo()
```
